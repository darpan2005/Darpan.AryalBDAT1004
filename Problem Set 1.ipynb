```python
#Question 3

using System;

public class Program
{
    public static void Main()
    {
        // Declare variables to store the lengths of the triangle sides
        double a, b, c;

        // Prompt the user to enter the length of side a and validate the input
        Console.Write("Enter the length of side a: ");
        if (double.TryParse(Console.ReadLine(), out a))
        {
            // Prompt the user to enter the length of side b and validate the input
            Console.Write("Enter the length of side b: ");
            if (double.TryParse(Console.ReadLine(), out b))
            {
                // Prompt the user to enter the length of side c and validate the input
                Console.Write("Enter the length of side c: ");
                if (double.TryParse(Console.ReadLine(), out c))
                {
                    // Calculate the area of the triangle using the provided sides
                    double area = CalculateTriangleArea(a, b, c);

                    // Check if the calculated area is valid
                    if (area >= 0)
                    {
                        Console.WriteLine("The area of the triangle is: " + area);
                    }
                    else
                    {
                        Console.WriteLine("Invalid input. Please enter valid numerical values for the sides.");
                    }
                }
                else
                {
                    Console.WriteLine("Invalid input. Please enter a valid numerical value for side c.");
                }
            }
            else
            {
                Console.WriteLine("Invalid input. Please enter a valid numerical value for side b.");
            }
        }
        else
        {
            Console.WriteLine("Invalid input. Please enter a valid numerical value for side a.");
        }
    }

    // Function to calculate the area of a triangle given its sides
    public static double CalculateTriangleArea(double a, double b, double c)
    {
        if (a > 0 && b > 0 && c > 0)
        {
            double s = (a + b + c) / 2;
            double area = Math.Sqrt(s * (s - a) * (s - b) * (s - c));
            return area;
        }
        else
        {
            return -1; // Indicates invalid input
        }
    }
}
```

```python
#Question 4

using System;

public class Program
{
    public static void Main()
    {
        // Prompt the user to input the number of elements to be stored in the array
        Console.Write("Input the number of elements to be stored in the array: ");
        int n = Convert.ToInt32(Console.ReadLine());

        // Declare arrays to store the input elements, even elements, and odd elements
        int[] arr = new int[n];
        int[] evenArr = new int[n];
        int[] oddArr = new int[n];
        
        // Initialize counters for even and odd elements
        int evenCount = 0;
        int oddCount = 0;

        // Prompt the user to input the elements of the array
        Console.WriteLine("Input {0} elements in the array:", n);
        for (int i = 0; i < n; i++)
        {
            Console.Write("element - {0} : ", i);
            arr[i] = Convert.ToInt32(Console.ReadLine());
        }

        // Separate even and odd elements into different arrays and update counters
        for (int i = 0; i < n; i++)
        {
            if (arr[i] % 2 == 0)
            {
                evenArr[evenCount] = arr[i];
                evenCount++;
            }
            else
            {
                oddArr[oddCount] = arr[i];
                oddCount++;
            }
        }

        // Display the even elements
        Console.WriteLine("The Even elements are:");
        for (int i = 0; i < evenCount; i++)
        {
            Console.Write(evenArr[i] + " ");
        }

        // Display the odd elements
        Console.WriteLine("\nThe Odd elements are:");
        for (int i = 0; i < oddCount; i++)
        {
            Console.Write(oddArr[i] + " ");
        }
    }
}
```

```python
#Question 6

def pig(word):
    # Convert the word to lowercase to handle uppercase characters
    word = word.lower()
    
    # Define a list of vowels
    vowels = ['a', 'e', 'i', 'o', 'u']
    
    # Check if the word starts with a vowel
    if word[0] in vowels:
        return word + 'way'
    else:
        # Find the index of the first vowel in the word
        index = -1
        for i, char in enumerate(word):
            if char in vowels:
                index = i
                break
        
        # Rearrange the word to create the pig-Latin form
        return word[index:] + word[:index] + 'ay'

# Ask the user to input a word
user_input = input("Enter a word: ")

# Call the pig function with the user's input and print the result
result = pig(user_input)
print("Pig-Latin:", result)
```

```python
#Question 7

def bldcount(blood):
    # Create a dictionary to store the count of each blood type
    blood_type_count = {'A': 15, 'B': 1, 'AB': 13, 'O': 14, 'OO': 0}

    try:
        # Open and read the file
        with open(blood, 'r') as file:
            lines = file.readlines()

        # Count the occurrences of each blood type
        for line in lines:
            blood_type = line.strip()  # Remove leading/trailing whitespace
            if blood_type in blood_type_count:
                blood_type_count[blood_type] += 1

        # Print the results
        for blood_type, count in blood_type_count.items():
            if count == 1:
                print(f'There is one patient of blood type {blood_type}.')
            else:
                print(f'There are {count} patients of blood type {blood_type}.')

    except FileNotFoundError:
        print(f"The file '{blood}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Call the bldcount function with the file name
bldcount('bloodtype1.txt')
```

```python
#Question 9

1) Trying to add incompatible variables, as in adding 6 + 'a':

==> Type of Exception: TypeError
==> Explanation: This will result in a TypeError because you cannot perform addition between an integer (6) and a string ('a') without explicit conversion.

2) Referring to the 12th item of a list that has only 10 items:

==> Type of Exception: IndexError
==> Explanation: Attempting to access an index that is out of the bounds of a list will raise an IndexError.

3)Using a value that is out of range for a function's input, such as calling math.sqrt(-1.0):

==> Type of Exception: ValueError (for math.sqrt(-1.0))
==> Explanation: The math.sqrt function expects a non-negative number as input, so passing a negative number will result in a ValueError. In some cases, it may also raise a math domain error.

4) Using an undeclared variable, such as print(x) when x has not been defined:

==> Type of Exception: NameError
==> Explanation: Using an undeclared variable (a variable that has not been defined before it is used) will raise a NameError.

5) Trying to open a file that does not exist, such as mistyping the file name or looking in the wrong directory:

==> Type of Exception: FileNotFoundError
==> Explanation: Attempting to open a file that does not exist, either due to a mistyped filename or the file being located in a different directory, will result in a FileNotFoundError.
```

```python
#Question 10

def frequencies(text, letters_to_count):
    char_count = {}  # Create an empty dictionary to store character frequencies

    # Convert the input text to lowercase for case-insensitive counting
    text = text.lower()

    # Iterate through the characters in the text
    for char in text:
        if char in letters_to_count:
            if char in char_count:
                char_count[char] += 1
            else:
                char_count[char] = 1

    return char_count

# Allow the user to input the letters they want to count
user_input = input("Enter the letters to count (e.g., 'abcdefghijklmnopqrstuvwxyz'): ")
letters_to_count = user_input.lower()

# Allow the user to input the text to analyze
text = input("Enter the text to analyze: ")

result = frequencies(text, letters_to_count)
print(result)
```
